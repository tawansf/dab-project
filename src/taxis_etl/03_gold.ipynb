```python
from pyspark.sql.functions import sum, avg, count, round, desc

source_table = "workspacecursodatabricks.taxis.silver_trips"
target_table = "workspacecursodatabricks.taxis.gold_analise_pagamento"

df =  spark.read.table(source_table)

df_gold = df.groupBy("cep_origem").agg(
    count("*").alias("total_corridas"),
    round(sum("valor_da_corrida"), 2).alias("faturamento_total"),
    round(avg("valor_da_corrida"), 2).alias("ticket_medio")
).orderBy(desc("faturamento_total"))

df_gold.write.format('delta').mode("overwrite").option('overwriteSchema', 'true').saveAsTable(target_table)

```
