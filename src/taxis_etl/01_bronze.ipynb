```python
from pyspark.sql.functions import current_timestamp, lit

source_table = "samples.nyctaxi.trips"
target_table = "workspacecursodatabricks.taxis.bronze_trips"

df_raw_data = spark.table(source_table).limit(5000)

df_bronze = df_raw_data.withColumn("ingestion_timestamp", current_timestamp()).withColumn("data_source", lit("nyc_taxi"))

df_bronze.write.format('delta').mode("overwrite").option('overwriteSchema', 'true').saveAsTable(target_table)
```
